```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(dataset):
    newdfvif=pd.DataFrame()
    newdfvif["Features"]=dataset.columns
    newdfvif["vif"]=[variance_inflation_factor(dataset.values,i)
                 for i in range(len(dataset.columns))]
    return(newdfvif)
```
